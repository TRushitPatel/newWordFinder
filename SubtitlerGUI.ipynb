```python
#gui
from tkinter import *
#selecting default path
from os import path
#file selection
from tkinter import filedialog
#progress bar
from tkinter.ttk import Progressbar
from tkinter import ttk
#color selection radio buttons
from tkinter.ttk import *

 
window = Tk()
 
window.title("Subtitler")
 
window.geometry('400x200')
 

#1 subtitle srt file selection
def subtitleFileSelection():
    subtitleFile = filedialog.askopenfilename(filetypes = (("subtitle","*.srt"),("all files","*.*")))
    lbl1.configure(text=subtitleFile)
    bar['value'] = 10
    
btn1 = Button(window, text="Select subtitle file", command=subtitleFileSelection)
btn1.grid(column=1, row=1)

lbl1 = Label(window, text="")
lbl1.grid(column=2, row=1)

    
#2 common word file selection
def commonWordsFileSelection():
 
    commonWordFile = filedialog.askopenfilename(filetypes = (("commonWordsFile","*.txt"),("all files","*.*")))
    lbl2.configure(text=commonWordFile)
    bar['value'] = 20
    
btn2 = Button(window, text="Select commonWord file", command=commonWordsFileSelection)
btn2.grid(column=1, row=2)

lbl2 = Label(window, text="")
lbl2.grid(column=2, row=2)


#3 radio button gui to select color
selected = IntVar()
 
rad1 = Radiobutton(text='First', value=1, variable=selected)
rad2 = Radiobutton(text='Second', value=2, variable=selected)
rad3 = Radiobutton(text='Third', value=3, variable=selected)
rad1.grid(column=1, row=3)
rad2.grid(column=2, row=3)
rad3.grid(column=3, row=3)

lbl3 = Label(window, text="")
lbl3.grid(column=5, row=3)

def colorSelection():
    lbl3.configure(text=selected.get())
    bar['value'] = 20
    
colorSelectionBtn = Button(window, text="done", command=colorSelection)
colorSelectionBtn.grid(column=4, row=3)
 

#4 progress bar  
style = ttk.Style()
style.theme_use('default')
style.configure("black.Horizontal.TProgressbar", background='black')
bar = Progressbar(window, length=200, style='black.Horizontal.TProgressbar')
bar['value'] = 0
bar.grid(column=1, row=4)

window.mainloop()
```
